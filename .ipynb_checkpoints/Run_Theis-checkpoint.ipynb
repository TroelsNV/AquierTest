## Dataset 1:
Confined aquifer

```python
# read dataset in the file "1.dat" located in the subfolder "data"
PW_rad = 0.01 # this radius is not meaningfull, and it is only for visualization
Q = 6/(60*60)
```

```python
S = 0.1
T = 2.2e-4*50
t = np.logspace(0, 6, num=100)
O1_sim = at.T1945(Q, T, S, 200, t)
O2_sim = at.T1945(Q, T, S, 150, t)
O3_sim = at.T1945(Q, T, S, 50, t)
O4_sim = at.T1945(Q, T, S, 10, t)
```

```python
f1, ax = plt.subplots(figsize=(15,5))

# plot the datasets as a function of t
#ax.loglog(data1['O1'][:,0],data1['O1'][:,1]*Q/T,'r*', label = 'O1')
#ax.loglog(data1['O2'][:,0],data1['O2'][:,1]*Q/T,'b*', label = 'O2')

# plot the solution as a function of t
ax.loglog(t,O1_sim,'r-', label = '200 m')
ax.loglog(t,O2_sim,'b-', label = '150 m')
ax.loglog(t,O3_sim,'k-', label = '50 m')
ax.loglog(t,O4_sim,'y-', label = '10 m')

# make the figure more readable
ax.set_xscale('log')
ax.grid(True,which="both")

#ax.set_title(file_name1)
ax.legend()
ax.set_ylim((1e-3, 1))
ax.set_xlim((1e3, 1e6))
ax.set_xlabel('t [s]')
ax.set_ylabel('s√¶nkning [m]')
```

## dataset 6:

```python
# read dataset in the file "1.dat" located in the subfolder "data"
file_name6 = os.path.join('data','6.dat')
data6 = read_data(file_name6)
PW_rad = 0.1 # this radius is not meaningfull, and it is only for visualization
Q = 1
b = 10
```

```python
data6['O1_dist'], data6['O2_dist']
```

```python
# Theis Solution:
Sy, S, T = 0.10308764079580357, 5e-3, 0.4165157203870851
PW_sim1 = at.T1945(Q, T, Sy, PW_rad, data6['O1'][:,0])
O1_sim1 = at.T1945(Q, T, Sy, data6['O1_dist'], data6['O1'][:,0])
O2_sim1 = at.T1945(Q, T, Sy, data6['O2_dist'], data6['O2'][:,0])

PW_sim2 = at.T1945(Q, T, S, PW_rad, data6['O1'][:,0])
O1_sim2 = at.T1945(Q, T, S, data6['O1_dist'], data6['O1'][:,0])
O2_sim2 = at.T1945(Q, T, S, data6['O2_dist'], data6['O2'][:,0])
```

### plot the data:

```python
f1, ax = plt.subplots(figsize=(15,5))

# plot the datasets as a function of t
#ax.loglog(data1['O1'][:,0],data1['O1'][:,1]*Q/T,'r*', label = 'O1')
#ax.loglog(data1['O2'][:,0],data1['O2'][:,1]*Q/T,'b*', label = 'O2')

ax.loglog(data6['O1'][:,0],data6['O1'][:,1],'r*', label = 'O1')
ax.loglog(data6['O2'][:,0],data6['O2'][:,1],'b*', label = 'O2')


# plot the solution as a function of t
ax.loglog(data6['O1'][:,0],O1_sim1,'r-', label = 'O1 late')
ax.loglog(data6['O2'][:,0],O2_sim1,'b-', label = 'O2 late')

# plot the solution as a function of t
ax.loglog(data6['O1'][:,0],O1_sim2,'r--', label = 'O1 early')
ax.loglog(data6['O2'][:,0],O2_sim2,'b--', label = 'O2 early')


# make the figure more readable
ax.set_xscale('log')
ax.grid(True,which="both")

ax.set_title(file_name5)
ax.legend()
ax.set_ylim((1e-5, 10.0))
ax.set_xlabel('t')
ax.set_ylabel(data1['columns'][1])
```
